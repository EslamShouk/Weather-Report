# Data Processing

```python
df["Date/Time"] = pd.to_datetime(df["Date/Time"])
df.dtypes
```

```python
# Add Month Column
df["Month"] = df["Date/Time"].dt.month_name()
df
```

```python
# Create a function to calculate time of date
def time_of_day(hour):
    if 6 <= hour < 12:
        return "Morning"
    elif 12 <= hour < 18:
        return "Afternoon"
    elif 18 <= hour < 24:
        return "Evening"
    else:
        return "Night"
```

```python
# apply the function
df['Time_of_Day'] = df['Date/Time'].dt.hour.apply(time_of_day)
df
```

```python
# Create a function to classification the Wind Speed
def classify_wind(speed):
    if speed < 10:
        return "Calm"
    elif speed < 30:
        return "Breezy"
    elif speed < 50:
        return "Windy"
    else:
        return "Stormy"
```

```python
# apply the function
df['Wind_Category'] = df['Wind Speed_km/h'].apply(classify_wind)
df
```

```python
# Create a function to classification the Temperature
def classify_temp(temp):
    if temp <= 0:
        return "Cold"
    elif temp <= 20:
        return "Mild"
    elif temp <= 30:
        return "Warm"
    else:
        return "Hot"
```

```python
# apply the function
df["temp_Category"]= df["Temp_C"].apply(classify_temp)
df
```

```python
# Create a function to classification the Visibility
def classify_Visibility(vis):
    if vis <= 2:
        return "Very Low"
    elif vis <= 5:
        return "Low"
    elif vis <= 15:
        return "Med"
    else:
        return "High"
```

```python
df["Visib_Category"] = df["Visibility_km"].apply(classify_Visibility)
df
```

```python
# Create a function to classification date
def get_season(date):
    if pd.Timestamp(date) >= pd.Timestamp("2012-12-21") or pd.Timestamp(date) < pd.Timestamp("2012-03-21"):
        return "Winter"  
    elif pd.Timestamp(date) >= pd.Timestamp("2012-03-21") and pd.Timestamp(date) < pd.Timestamp("2012-06-21"):
        return "Spring"  
    elif pd.Timestamp(date) >= pd.Timestamp("2012-06-21") and pd.Timestamp(date) < pd.Timestamp("2012-09-23"):
        return "Summer"  
    elif pd.Timestamp(date) >= pd.Timestamp("2012-09-23") and pd.Timestamp(date) < pd.Timestamp("2012-12-21"):
        return "Fall"    
```

```python
df["Season"] = df["Date/Time"].apply(get_season)
df
```

```python
df["Season"].value_counts()
```

```python
monthly_summary = df.groupby('Month').agg({
    'Temp_C': 'mean',
    'Visibility_km': 'mean',
    'Wind Speed_km/h': 'mean'
}).reset_index()
```
